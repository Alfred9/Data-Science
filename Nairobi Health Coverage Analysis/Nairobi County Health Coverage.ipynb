#### Healthcare Facilities Types Distribution Across subcounties

```python

# Get unique facility types from the "Type" column
service_types = nairobi_df["Type"].unique()

# Plot distribution for each facility type
for service_type in service_types:
    # Filter data for the current facility type
    service_data = nairobi_df[nairobi_df["Type"] == service_type]
    
    # Group the filtered data by sub-county and calculate the count of the current facility type
    service_distribution = service_data.groupby("SubCounty").size().reset_index(name="Count")
    
    # Plot the distribution of the current facility type across sub-counties
    plt.figure(figsize=(12, 6))
    sns.barplot(x="SubCounty", y="Count", data=service_distribution)
    plt.title(f"{service_type} Distribution Across Sub-Counties")
    plt.xlabel("Sub-County")
    plt.ylabel("Number of Facilities")
    plt.xticks(rotation=45)
    plt.tight_layout()

    #save the plot
    plt.savefig(f"{service_type.lower().replace(' ', '_')}_distribution.png")
    plt.show()
```

### Merge  the 2 datasets

```python
# load both the nairobi health facility dataset and the population datase
df1 = pd.read_csv('Population_data.csv')
df2 = pd.read_csv('nairobi_data.csv')
```

 ### data preprocessing  to standardize and merge the two datasets

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process


# Standardize subcounty names by removing extra spaces and converting to lowercase
df1['SubCounty'] = df1['SubCounty'].str.strip().str.lower()
df2['SubCounty'] = df2['SubCounty'].str.strip().str.lower()

# Define a function to find matching subcounties
def find_matching_subcounty(subcounty, subcounty_list):
    """
    Finds the best matching subcounty in a list using fuzzy matching.
    Returns None if no match is found.
    """
    matches = process.extractOne(subcounty, subcounty_list, scorer=fuzz.token_set_ratio)
    if matches[1] >= 73:  # Set a threshold for fuzzy matching
        return matches[0]
    else:
        return None

# Create a mapping between subcounties in df1 and df2
subcounty_mapping = {}
for subcounty in df1['SubCounty'].unique():
    match = find_matching_subcounty(subcounty, df2['SubCounty'].unique())
    if match:
        subcounty_mapping[subcounty] = match

# Update subcounty names in df1 using the mapping
df1['subcounty_mapped'] = df1['SubCounty'].map(subcounty_mapping)

# Merge the datasets on the standardized 'subcounty_mapped' column
merged_df = pd.merge(df2, df1, how='inner', left_on='SubCounty', right_on='subcounty_mapped')

# Drop redundant columns
merged_df.drop(['subcounty_mapped', 'SubCounty_y'], axis=1, inplace=True)


merged_df = merged_df.rename(columns={'SubCounty_x':'SubCounty'})


# Display the updated DataFrame
merged_df
```

```python
# Select 'SubCounty' and 'Total' columns from the merged dataset
unique_subcounty_total = merged_df.drop_duplicates(subset='SubCounty')[['SubCounty', 'Total']]

# Display the result
unique_subcounty_total
```

### Population Density

```python
plt.figure(figsize=(12, 6))
plt.bar(unique_subcounty_total['SubCounty'], unique_subcounty_total['Total'], color='skyblue')
plt.title('Total Population Across Sub-Counties')
plt.xlabel('Sub-County')
plt.ylabel('Total Population')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('Population Across Sub-Counties.png')
plt.show()
```

```python
# Calculate the number of facilities per sub-county
facility_counts = merged_df['SubCounty'].value_counts().reset_index()
facility_counts.columns =['SubCounty', 'facilities_no']

facility_counts
```

```python
# Merge the unique_subcounty_total DataFrame with the facility_counts DataFrame
new_df = pd.merge(unique_subcounty_total, facility_counts, on='SubCounty')

# Calculate the number of persons per facility
new_df['Persons_per_Facility'] = new_df['Total'] / new_df['facilities_no']

# Display the resulting DataFrame
print(new_df)
```

### Number of Persons per Facility Distribution:

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by the number of persons per facility
new_df1 = new_df.sort_values(by='Persons_per_Facility', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
plt.bar(new_df1['SubCounty'], new_df1['Persons_per_Facility'], color='skyblue')
plt.title('Number of Persons per Facility by Sub-County')
plt.xlabel('SubCounty')
plt.ylabel('Number of Persons per Facility')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('Persons per Facility in Each Sub-County.png')
plt.show()
```
